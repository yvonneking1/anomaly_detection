```python
colnames=['timestamp', 'times', 'request_agent', "student_id", "cohort", "ip"]
df = pd.read_csv("curriculum-access.txt",
                 engine='python',
                 header=None,
                 index_col=False,
                 names=colnames,
                 sep=r'\s(?=(?:[^"]*"[^"]*")*[^"]*$)(?![^\[]*\])',
                 na_values='"-"')
```

```python
df["Timestamp"] = df["timestamp"] + " " + df["times"]
```
